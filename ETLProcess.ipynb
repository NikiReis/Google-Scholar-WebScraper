Data Transformation

```python
schema = pa.DataFrameSchema(
    columns={
        'Book': pa.Column(pa.String),
        'Authors': pa.Column(pa.String),
        'Jornal': pa.Column(pa.String, nullable = True),
        'Year': pa.Column(pa.Int, nullable = True),
        'Publisher': pa.Column(pa.String, nullable = True),
        'Link': pa.Column(pa.String, nullable = True)
    }
)
```

```python
schema.validate(df)
```
